### Ex1 α)

```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

Training_Set = sio.loadmat('Training_Set.mat')
X = Training_Set['X']
y = Training_Set['y'][:,0]
```

```python
#4a)
plt.scatter(X,y)
plt.show()
```

```python
#initialize with x0 as ones
X_poly = np.ones(len(X))

#loop for the powers 1-8 and add the column
for i in range(1,9):
    X_poly = np.column_stack([X_poly, X ** i])
```

```python
#b calculate the thetas using least squares
ls_thetas = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y            
#create a linspace of 10000 datapoints between 0 and 1
X_line = np.linspace(0,0.9,10000)

#for each of the 10000 datapoints, calculate the polynomial terms x**2, x**3 ... x**8
X_poly_line = np.ones(10000)
for i in range(1,9):
    X_poly_line = np.column_stack([X_poly_line, X_line ** i])
y_line = X_poly_line @ ls_thetas

#plot the results
plt.scatter(X,y, color = 'red', label = 'data points')
plt.plot(X_line, y_line, label = '8th degree polynomial fit')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.title('LS fit line')
plt.show()

print('thetas = ',  ls_thetas.reshape(-1,1))
```

#### c)
Since we want to ommit the intercept from the penalization, we will adjust our cost function so that it minimizes the following

$J(θ) = \sum_{i=1}^{N}(y_i - θ_0 - θ^Τx_i)^2 + λ\|θ\|^2$ where X is our polynomial dataset without the columns of 1s
First we will take the partial derivative  
wrt $θ_0$ and set it to 0  
$\frac{J(θ)}{θ_0} = 2\sum_{i=1}^{N}(y_i - θ_0 - θ^Τx_i) = 0$  
$\Rightarrow \sum_{i=1}^{N}(y_i - θ_0 - θ^Τx_i) = 0$  
$\Rightarrow \sum_{i=1}^{N}θ_0  - \sum_{i=1}^{N}y_0 + \sum_{i=1}^{N} θ^Τx_i = 0$  
$\Rightarrow Nθ_0  = \sum_{i=1}^{N}y_0 - \sum_{i=1}^{N} θ^Τx_i$  
$\Rightarrow θ_0  = \frac{\sum_{i=1}^{N}y_0}{N} - \frac{\sum_{i=1}^{N} θ^Τx_i}{N}$  
$\Rightarrow θ_0  = \bar{y} - θ^Τ\bar{x}$  
subsituting $θ_0$ into our cost function we get
$J(θ) = \sum_{i=1}^{N}(y_i - \bar{y} + θ^Τ\bar{x} - θ^Τx_i)^2 + λ\|θ\|^2$  
$\Rightarrow J(θ) = \sum_{i=1}^{N}[y_i - \bar{y}  - θ^Τ(x_i - \bar{x})]^2 + λ\|θ\|^2$ setting y centered as $y^{(c)}_i = y_i - \bar{y}$ and x centered as $x^{(c)}_i = x_i - \bar{x}$ we get  
$J(θ) = \sum_{i=1}^{N}(y^{(c)}_i  - θ^Τx^{(c)}_i)^2 + λ\|θ\|^2$  
now we can take the derivative of the above wrt $θ$ set it to 0 and solve using the matrix form as in 1b)
$θ = (Χ_c^ΤΧ_c + λΙ)^{-1}Χ_c^Τy_c$  
After finding all $θ$ we can calculate $θ_0$

```python
# c)
#center Xs
X_offset = X_poly[:,1:].mean(axis=0)
X_c = X_poly[:,1:] - X_offset

#center the ys
y_offset = y.mean(axis=0)
y_c = y - y_offset
```

```python
plt.figure(figsize=(16, 8))
cmap = plt.get_cmap('Blues_r')
colors = iter(cmap(np.linspace(0,1,20)))

#initialize a list to store thetas for every iteration
ridge_thetas_list = []
for lambda_ in np.arange(0,0.0002,0.00002):
    #calculate thetas using the centered X and y, ommitting the bias from the regularization
    ridge_thetas = np.linalg.inv(X_c.T @ X_c + lambda_ * np.identity(X_c.shape[1])) @ X_c.T @ y_c
    #calculate the bias and append
    ridge_theta0 = y_offset - X_offset @ ridge_thetas
    ridge_thetas = np.hstack([ridge_theta0, ridge_thetas])
    #append ridge_thetas to the list
    ridge_thetas_list.append(ridge_thetas)
    #calculate y_hat
    y_hat = X_poly_line @ ridge_thetas 
    plt.plot(X_line, y_hat, label = 'λ='+str(lambda_), color = next(colors))
plt.scatter(X,y, color = 'black', label = 'data points')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Ridge Regression varying regularization parameter')
plt.legend()
plt.show()
```

```python
#d)
from sklearn.linear_model import Lasso

plt.figure(figsize=(16, 8))
cmap = plt.get_cmap('Blues_r')
colors = iter(cmap(np.linspace(0,1,20)))

#initialize a list to store thetas for every iteration
lasso_thetas_list = []
for lambda_ in np.arange(0,0.0002,0.00002):
    #initialize a Lasso regressor
    reg = Lasso(alpha=lambda_)
    #fit the model. Here we omit the intercept column, since it will be calculated from scipy
    reg.fit(X_poly[:,1:], y)
    lasso_thetas = np.hstack([reg.intercept_, reg.coef_])
    #append lasso_thetas to the list
    lasso_thetas_list.append(lasso_thetas)
    #calculate y_hat
    y_hat = X_poly_line @ lasso_thetas 
    plt.plot(X_line, y_hat, label = 'λ='+str(lambda_), color = next(colors))
plt.scatter(X,y, color = 'black', label = 'data points')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.title('Lasso Regression varying regularization parameter')
plt.show()
```

```python
#e)
fig, axs = plt.subplots(3,3, figsize= (16,10))
axs = axs.ravel()
fig.subplots_adjust(hspace = 0.5)
ridge_df = pd.DataFrame(ridge_thetas_list, columns=['theta'+str(i) for i in range(0,9)])
lasso_df = pd.DataFrame(lasso_thetas_list, columns=['theta'+str(i) for i in range(0,9)])
for i,theta in enumerate(ridge_df.columns):
    axs[i].plot(np.arange(0,0.0002,0.00002), ridge_df[theta], label = 'ridge')
    axs[i].plot(np.arange(0,0.0002,0.00002), lasso_df[theta], label = 'lasso')
    axs[i].legend()
    axs[i].title.set_text('$θ$'+str(i) + ' coefficient over $λ$')
    axs[i].locator_params(axis='x', nbins=6)
```

Each plot above depicts the coefficient path of each $θ$ over the increase of the regularization parameter $λ$. There are 2 paths in each plot which visualize the effect of ridge regression L2 penalization and lasso regression L1 penalization. The common effect of the two types of regression is that they shrink those coefficients $θ$ which result to big variance. Coefficients like $θ_0$ and $θ_1$ are not fully suppressed yet for the particular range of lambdas, since our model could be approximated fairly well linearly. On the other hand, higher grade polynomial parameters are suppressed quicker to 0 or a value really close to it.

Additionally, we can also observe how different penalizers affect the coefficients. Ridge tends to shrink the coefficients more gradually and for small lambdas they might not reach 0 exactly. On the other hand Lasso penalizes the coefficients more abruptly, reducing them to 0 exactly.

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

#5a) instead of a loop we will implement this in a vectorized fashion
np.random.seed(123)
#dimensions: 50 datasets with 30 rows each
n=30
d=50

#X is a matrix where each column is a dataset. The shape is (30,50)
X = np.random.uniform(0,20,(n,d))

#generate noise for each dataset, with 0 means, 64 variance and 30 rows. The shape is (30,50)
noise = np.random.multivariate_normal(np.zeros(50),np.identity(50)*64,n)

#scalar multiplication with theta=2 and add the noise. The shape is the same so no issues should arise
y = X * 2 + noise

#calculate the least squares estimates as well as the MSE
thetas_LS = np.multiply(X,y).sum(axis=0) / (np.multiply(X,X).sum(axis=0))
MSE_LS = np.sum((thetas_LS-2)**2) / len(thetas_LS)

#initialize empty lists for keeping track of the variance and bias of the MSE of the estimator
bias_arr = []
variance_arr = []
MSE_arr = []
lambdas = np.arange(0,200,0.1)

for lambda_ in lambdas:
    #calculate the thetas subject to regularization
    thetas = np.multiply(X,y).sum(axis=0) / (np.multiply(X,X).sum(axis=0) + lambda_)
    #calculate the variance of the thetas
    variance = ((thetas - thetas.mean())**2).mean()
    variance_arr.append(variance)
    #calculate the bias of the thetas
    bias = (thetas.mean() - 2)**2
    bias_arr.append(bias)
    MSE = np.sum((thetas-2)**2) / len(thetas)
    MSE_arr.append(MSE)
    
df = pd.DataFrame({
    'MSE' : MSE_arr
    ,'bias' : bias_arr
    ,'variance' : variance_arr
    ,'lambda_' : lambdas
})

plt.plot(df.lambda_, df.MSE, label='MSE($θ$) s.t. regularization')
plt.hlines(MSE_LS, min(lambdas), max(lambdas), color = 'red', label = 'MSE of LS Estimator')
plt.xlabel('$λ$')
plt.ylabel('MSE')
plt.title('Comparing Least squares estimator with Regularized Estimators')
plt.legend()
plt.show()

lambda_lower = round(df[df.MSE <= MSE_LS].lambda_.min(),2)
lambda_higher = round(df[df.MSE <= MSE_LS].lambda_.max(),2)
print('For lambda between {} and {} the MSE of the estimator is smaller than the LS MSE'.format(lambda_lower, lambda_higher))
```

We can see that for some regularized estimations, the MSE of the estimator is lower than the MSE of the unbiased LS estimator. In effect, by regularizing, we are reducing the variance of the estimator at the cost of introducing some bias. Neverthelesess, the reduced variance assures us that there is not high volatility and dependency on the data. As a result, our model has higher chances to perform better during the prediction phase. In the next plot we can see how the 2 components of the MSE, the bias and the variance evolve over the increase of lambda

```python
data_plot = df[df.lambda_ <= 500]
fig, ax1 = plt.subplots(figsize=(6,4))

color = 'tab:red'
ax1.set_xlabel('$λ$')
ax1.set_ylabel('bias', color=color)
ax1.plot(data_plot.lambda_, data_plot.bias, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.title.set_text('Dual axis of bias-variance (different scale)')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('variance', color=color)  # we already handled the x-label with ax1
ax2.plot(data_plot.lambda_, data_plot.variance, color=color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

As expected with the increase of the regularization parameter, bias is introduced and variance is reduced from our estimations.
