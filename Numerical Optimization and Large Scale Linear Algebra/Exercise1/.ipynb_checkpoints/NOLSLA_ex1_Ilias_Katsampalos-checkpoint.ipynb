# i) The Fibonacci Sequence
## a) Plotting

```python
def fib(n,c):
    '''
    Calculates n fibonacci numbers, distirbed by the factor c
    
    Parameters:
        n (int): The length of the Fibonacci Sequence
        c (float64): The disturbance factor
    
    Returns:
        res (list): the n-length fibonacci sequence
    '''
    #initializing
    res = np.zeros(n,dtype='double')
    res[0] = 1
    res[1] = 1
    #sequence calculation
    for i in range(2,n):
        number = c * res[i-1] + res[i-2]
        res[i] = number
    return res

n = 100
x = list(range(n))
#calculate the sequence a, with the undisturbed fibonacci numbers
a = fib(n,1)
#calculate the sequence p, with the disturbed fibonacci numbers
c = 1 + np.sqrt(3)/100
p = fib(n,c)

#retrieve the machine epsilon for single and double
e_machine_double = np.finfo('double').eps
e_machine_single = np.finfo('single').eps
```

```python
from matplotlib import pyplot as plt
fig, axs = plt.subplots(figsize=(8,6))
plt.semilogy(x,a,label='c=1')
plt.semilogy(x,p,label='$c=1+\sqrt{3}/100$')
plt.xlabel('n')
plt.hlines((1/e_machine_double),0,n,label='1/$e_{machine-double}$', linestyles='--')
plt.hlines((1/e_machine_single),0,n,label='1/$e_{machibe-single}$',linestyles='dotted')
plt.legend()
plt.show()
```

## b) Descending the Fibonacci and error analysis

We have that $a_{k+1} = a_k + a_{k-1} \Rightarrow a_{k-1}=a_{k+1}-a_k$

```python
c=1 + np.sqrt(3)/100
n = 100
a = fib(n,1)
p = fib(n,c)
```

```python
def reverse_fib(fib_sequence, c):
    ''' 
    Given a list of fibonacci numbers, start descending 
    for each one until reaching the first number of the sequence
    
    Parameters:
        fib_sequence (list): A sequence of Fibonacci numbers
        c (float64): The disturbance factor 
        
    Rerutns:
        n_0 (np.array): Array containing the 1st fibonacci number
                        for each descend, e.g. n_0[18] is the approximation
                        of the 1st fibonacci number staring from the 18th number
    '''
    
    #Initialization of an array, which holds the 1st number of each descent
    n_0 = np.zeros(len(fib_sequence), dtype='double')
    n_0[0] = 1
    n_0[1] = 1
    #For each number of the fibonacci sequence, starting from the end
    for n in range(len(fib_sequence), 2, -1):
        #Initilize intermediate array
        res = np.zeros(n, dtype='double')
        res[n-1] = fib_sequence[n-1]
        res[n-2] = fib_sequence[n-2]
        #calculate the descending sequence
        for i in range(n-2,0,-1):
            res[i-1] = res[i+1] - res[i] * c
        #append the first number to the initial array
        n_0[n-1] = res[0]
    return n_0

a_0 = reverse_fib(a,1)
```

After calculating all a_0, we calculate the absolute relative error with the number 1 and plot them.

```python
#define a lambda function to calculate the abs rel error.
absolute_relative_error_scalar = lambda x, x_hat: abs(x - x_hat) / x
```

```python
a_errors = absolute_relative_error_scalar(1,a_0)
fig, axs = plt.subplots(1,2,figsize=(14,4))
fig.subplots_adjust(top=0.8)

axs[0].semilogy(list(range(1,n+1)), a_errors)
axs[0].set_xlabel('n')
axs[0].set_ylabel('$log$ error')
axs[0].title.set_text("All Numbers")

axs[1].semilogy(list(range(70,n)), a_errors[70:])
axs[1].set_xlabel('n')
axs[1].set_ylabel('$log$ error')
axs[1].title.set_text("Zoomed in error area")
fig.suptitle("Log Abs. Rel. Error between Fibonacci's 1st number and approximation")
plt.show()
```

We can start observing deviations from the first fibonacci number `1`, when starting to calculate back fibonacci numbers starting from the `78th` number. Additionally, the greater the number `n` from which we start the calculation back, the greater the log of the error.

Now, why do we observe errors starting from the 78th fibonacci number? (counter starts from 0). Using the theory we have:
* for a given exponent `e` we can write $2^{k-1}$ different numbers where k is our precision
* between $2^e$ and $2^{e+1}$ there is a space of $2^e$ 
* since in that space we can write up to $2^{k-1}$ numbers, the gap between them will be $\frac{2^e}{2^{k-1}}=2^{e-k+1}$
* the machine epsilon is normalized to the 1 exponent

For example between $2^0$ and $2^1$ we have a gap size of:

```python
2 **(0-53+1)
```

which is equal to the machine epsilon for the double type

```python
e_machine_double
```

We also know that in each exponent, the gap size doubles in size. For example for the next group of exponents  
$2^1$ and $2^2$ we have:

```python
2 **(1-53+1)
```

in effect this is twice the machine epsilon

```python
2 * e_machine_double
```

Now in the case of the undisturbed fibonacci, we are working with integers. So, as long as the gap size < 1, we should not face any issues with precision. Issues will arise when we will have to calculate numbers where the gap size > 1. We have already stated that $gap size = \frac{2^e}{2^{k-1}}$. Reforming the equation and substutiting $gap size = 1$ we have $2^e = 2^{k-1} \Rightarrow e = 52$. In effect this means that all numbers that can be represented between $2^{52}$ and $2^{53}$ are integers with a spacing of 1.

Now recall that the fibonacci number where we started noticing deviations was the 78th (counter starts from 0).

```python
a[77] < 2**53 < a[78]
```

From the above boolean expression, we can see that we cannot store precisely the 78th number, since we have used all 53 bits of the mantissa. Between $2^{53}$ and $2^{54}$ we have the following gap size.

```python
2 ** (53 - 53 + 1)
```

In effect, in the range [$2^{53}$ and $2^{54}$) we can store only even numbers. Descending the fibonacci sequence from that number will introduce precision issues in our calculation.

## c) Descending the Disturbed Fibonacci and error analysis
Now we calculate the erros using the disturbed sequence `p`.

```python
p_0 = reverse_fib(p,c)
p_errors = absolute_relative_error_scalar(1,p_0)
fig, axs = plt.subplots(figsize=(6,4))
fig.subplots_adjust(top=0.8)

axs.semilogy(list(range(1,n+1)), p_errors)
axs.set_xlabel('n')
axs.set_ylabel('$log$ error')
axs.title.set_text("Log Abs. Rel. Error between Fibonacci's 1st number and disturbed approximation")
plt.show()
```

Using the slightly disturbed `p` sequence by a factor of $1 + \frac{\sqrt{3}}{100}$, we can can observe errors much earlier in the sequence. Combinig the two approaches into a single plot we have.

```python
plt.semilogy(list(range(1,n+1)), a_errors, label='c=1')
plt.semilogy(list(range(1,n+1)), p_errors, label='$c=1+\sqrt{3}/100$')
plt.xlabel('n')
plt.ylabel('$log$ error')
plt.legend()
plt.title('Absolute errors for $a_0$, $p_0$ for un/distirbed fibonacci sequence')
plt.show()
```

Now that we have the complete picture, we can draw some conclusions regarding the errors. We can observe, that deviations from the first number of the Fibonacci sequence start to occur for higher values of `n` for the undisturbed approach than the disturbed approch. After the 78th Fibonacci number, the error of the two approaches converges in magnitude. As one can observe from the plot above, the most typical approach in terms of accuracy is the undisturbed approach, since with disturbed approach was more prone to produce an error, due to the factor `c`.

What follows is an attemp to predict the magnitude of the error:

Since we are using the disturbed fibonacci sequence, there is already a rounding error inherited by the use of the factor $c = 1 + \frac{\sqrt{3}}{100}$, which unavoidably is $10^{-17}$ since in double precision we have ~16 significants. Now, suppose that we start descending the fibonacci sequence from a number, whose magnitude is $10^{2}$. Since this number was calculated using the distrubance factor `c`, and since the exponent is 2, we have discarded 2 significands of the factor `c`. That means that the error at this point has a magnitude of $10^{-17+2}$. Because we want to reach the first fibonacci number, which is of magnitude $10^0$ from our starting point number which was $10^{2}$, with a sequence of substractions, due to the alignment of the exponents we will lose another 2 significands. Thus our error will be $10^{-17 + 2 * 2}$.

So rounding up, we can approximate the magnitude of the error using: $$error = 2 *  exponent - 17$$

To see if our assumption holds, we can plot the magnitude of the relative errors together with the exponent of the base number from which we start to descent.

```python
def magnitude(x):
    base10 = np.log10(abs(x))
    return (np.floor(base10))

given_magnitude = magnitude(p[6:]) #starting from when errors start to appear
error_magnitude = magnitude(p_errors[6:]) #starting from when errors start to appear
plt.scatter(given_magnitude, error_magnitude)
plt.title('Magnitude of Starting Fib number vs Magn of Relative Error')
plt.xlabel('Magnitude of Fibonacci number')
plt.ylabel('Magnitude of Relative Error')
plt.show()
```

So in fact, from the plot above we see the linear relationship that we described using theory. As one can observe, the relationship between the magnitude of the fibonacci number where the descent starts and the magnitude of the error seem to be related in a linear way. Next, we fit a linear regression in order to compare the coefficients with our initial assumption. 

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(given_magnitude.reshape(-1,1), error_magnitude)
print(f'Coefficient: {lr.coef_}, intercept: {lr.intercept_}')
```

So indeed, the coefficients of the linear regression match our assumptions using theory. Trying to interpret the results we can say the following about the magnitude of the error: `Given the magnitude x of the starting number of the descending sequence, which has already lost x significands, double them up since we need to descend to a magnitude of 0. Subtract ~-17 which is the inherent rounding error due to the disturbance factor c.

# ii) The Binomial Coefficient

```python
def _cast_float(x, dtype):
    if dtype == 'single':
        return np.float32(x)
    else:
        return np.float64(x)

def _cast_int(x, dtype):
    if dtype == 'single':
        return np.int32(x)
    else:
        return np.int64(x)
    
def choose(n,kappa, dtype='double'):
    n = _cast_int(n,dtype)
    a = _cast_int(1,dtype)
    for k in range(kappa):
        k = _cast_int(k, dtype)
        one = _cast_int(1, dtype) 
        a = _cast_float( (n-k)/(k+one) * a, dtype)
    return a
```

```python
def binomial_all_success(iterations, dtype='double'):
    n = iterations
    if dtype == 'single':
        res = np.zeros(n,dtype='float32')
    else:
        res = np.zeros(n,dtype='float64')
    for i in range(1,n+1):
        res[i-1] = choose(i,i,dtype)
    error = absolute_relative_error_scalar(1,res)
    return error
```

```python
n = 1000
single_errors = binomial_all_success(n,'single')
double_errors = binomial_all_success(n,'double')
```

```python
fig, axs = plt.subplots(1,2, figsize=(14,4), sharex=True)
fig.subplots_adjust(top=0.8)

axs[0].plot(list(range(n)), single_errors)
axs[0].set_xlabel('n')
axs[0].set_ylabel('Abs Rel Error')
axs[0].title.set_text('Single Precision')

axs[1].plot(list(range(n)), double_errors)
axs[1].set_xlabel('n')
axs[1].set_ylabel('Abs Rel Error')
axs[1].title.set_text('Double Precision')

fig.suptitle('Abs Rel Error between binomial coefficient w all successes and 1')
plt.show()
```

Here we do not observe the same magnitude of errors as in exercise `i`. The reason is that the computation $a_{n,k+1}=\frac{n-k}{k+1}a_{n,k}$ is driven  mostly by multiplication. For multiplication, we do not need to shift the exponents to the same magnitude thus no leading digits are lost. On the other hand, the exponents are added and the mantissas multiplied, thus we are only losing the trailing digits due to rounding without suffering from loss of significance. 

## b) The Expected Value  
Function for calculating the expected vaue.

```python
def exp_value(n, dtype='double'):
    #casting everything to the appropriate type
    a = _cast_float(1,dtype)
    s = _cast_float(0,dtype)
    one = _cast_float(1, dtype) 
    two = _cast_float(2, dtype)
    #initialize a list to return the a_n,k
    alphas = []
    for k in range(n+1):
        alphas.append(a)
        n = _cast_int(n,dtype)
        k = _cast_int(k, dtype)
        s += _cast_float( k * a, dtype)
        a = _cast_float( (n-k)/(k+one) * a, dtype)
    exp =  _cast_float( one / two ** n, dtype) * s
    return exp, alphas
```

Calculating and plotting absolute relative errors for single precision.

```python
import pandas as pd
n = 400
res_single = np.zeros(n)
alphas = []
for i in range(1,n+1):
    exp_, alphas_ = exp_value(i, 'single')
    res_single[i-1] = absolute_relative_error_scalar(exp_, i/2)
    alphas_ = list(zip([i]*len(alphas_), alphas_))
    alphas += (alphas_)
```

```python
fig, axs = plt.subplots(figsize=(8,5))
plt.plot(list(range(1,n+1)), res_single)
plt.title('Abs Rel Error, Mean of Binomial Distribution | Single Prec')
plt.xlabel('Tosses')
plt.ylabel('Error From Theoretical Mean')
plt.show()
```

We do not observe a large magnitude of errors for single precision. Although, we received an overflow warning when calculating the summation. Now plotting the the log of the value $a_{n,k}$ for different numbers of tosses we have:

```python
single_alphas = pd.DataFrame(alphas, columns=['tosses','combinations'])
single_alphas['successes'] = single_alphas.groupby('tosses').cumcount()

import seaborn as sns
fig, axs = plt.subplots(figsize=(10,6))
sample_alphas = single_alphas[single_alphas.tosses.isin(np.arange(0,200,20))]
g = sns.lineplot(sample_alphas.successes, sample_alphas.combinations, hue=sample_alphas.tosses, legend="full")
plt.hlines(np.finfo('single').max, 0, sample_alphas.successes.max(), linestyle='dashed', label='max single')
plt.legend(loc='lower right')
plt.title('Binmomial Coefficient for different number of tosses')
plt.ylabel('log combinations')
g.set_yscale('log')
```

Here we have also plotted above the binomial coefficient for different number of tosses, as well as the maximum single precision number that we can represent using the floating system. We observe that for large values of tosses, we can not represent the binomial term as it produces an overflow.

## c) The Bell Shape

```python
_, alphas_double = exp_value(40,'double')

sns.lineplot(range(len(alphas_double)), alphas_double)
plt.xlabel('successes')
plt.ylabel('combinations')
plt.show()
```

From a statistical point of view, the bimomial coefficient returns all the possible combinations that `k` successes can be formed in `n` number of bernoulli experiments. For example, in 40 tosses, 0 successes can occur only in single combination, where all bernoulli outcomes are `False`. The same applies for 40 successes, since all bernoulli outcomes must be `True`. Now the number of combinations for `k` successes increases exponentially up to a point where the number of successes is the same as the number of failures, where it reaches its maximum. Then the number of combinations starts decreasing again in a symmetric way, since in effect, we want to distribute the same number of failures as the number of successes on the opposite part of the symmetry. Here is a quick proof of the symmetry of the binomial coefficient.

$$
\begin{align*}
{n \choose k} &= {n \choose n-k} \\
\Rightarrow \frac{n!}{k!(n-k)!} &= \frac{n!}{(n-k)!(n-(n-k))!} \\
\Rightarrow \frac{n!}{k!(n-k)!} &= \frac{n!}{(n-k)!(n- n + k))!} \\
\Rightarrow \frac{n!}{k!(n-k)!} &= \frac{n!}{k!(n-k)!}
\end{align*}
$$

## d)
We will define a function. In order to understand better what are the dangers of this calculation, we will keep track of the outcome and plot the progress.

```python
def s(n, dtype='double'):
    n = _cast_int(n,dtype)
    a = _cast_int(1,dtype)
    s = 0
    running_sum = []
    for k in range(n+1):
        k = _cast_int(k, dtype)
        to_add = (-1)**k * np.sin(2*np.pi*np.sin(k/n)) * a
        s += to_add
        running_sum.append(to_add)
        #print(np.frexp(to_add))
        one = _cast_int(1, dtype) 
        a = _cast_float( (n-k)/(k+one) * a, dtype)
    return s, running_sum
```

We will perform a run for n=200

```python
res, running_sum = s(200)
print(f'result for n=100: {res}')
```

Plotting the running sum we have

```python
from matplotlib import pyplot as plt
plt.plot(np.arange(201),running_sum)
plt.xlabel('n')
plt.ylabel('running sum')
plt.title('running sum for n=200')
plt.show()
```

We can spot two weak points in this algorithm, namely:

* the interchangable addition of opposite sign numbers 
* the summation of different magnitudes, since in the beginning the summation contains lower values, in the middle larger values and in the end lower values again. 

Climbing to higher exponents, like 58 for example, it can lead to catastrophic cancellation, since in effect, we are subtracting almost similar numbers or add very small numbers to a big summation, making them dissapear due to the exponent alignment. This results to loss of significance which impacts directly following operations.

To alleviate ourselves from this pain-point, we can keep track of 2 partial sums, one for the positive values of $-1^k$ and one for the negative. Also, the order that these sums will be performed, will be from the lowest number to the greatest.

```python
def kahan_sum(to_add, summation, c):
    y = to_add - c
    t = summation + y
    c = (t - summation) - y
    summation = t
    return summation, c


def s_b(n, dtype='double'):
    n = _cast_int(n,dtype)
    a = _cast_int(1,dtype)
    pos_sum_first_half = 0
    pos_sum_second_half = 0
    neg_sum_first_half = 0
    neg_sum_second_half = 0

    c_pos_first_half = 0
    c_pos_second_half = 0
    c_neg_first_half = 0
    c_neg_second_half = 0
    for k in range(int(n/2)+1):
        k = _cast_int(k, dtype)
        to_add = (-1)**k * np.sin(2*np.pi*np.sin(k/n)) * a
        if to_add > 0:
            pos_sum_first_half, c_pos_first_half = kahan_sum(to_add, pos_sum_first_half, c_pos_first_half)
        else:
            neg_sum_first_half, c_neg_first_half = kahan_sum(to_add, neg_sum_first_half, c_neg_first_half)
        one = _cast_int(1, dtype)
        a = _cast_float((n-k)/(k+one) * a, dtype)

    a = _cast_int(1, dtype)
    to_add = (-1) ** n * np.sin(2 * np.pi * np.sin(n / n)) * a
    if to_add > 0:
        pos_sum_second_half += to_add
    else:
        neg_sum_second_half += to_add

    for k in range(n-1, int(n / 2), -1):
        a = _cast_float((k + 1) / (n - k) * a, dtype)
        k = _cast_int(k, dtype)
        to_add = (-1) ** k * np.sin(2 * np.pi * np.sin(k / n)) * a
        if to_add > 0:
            pos_sum_second_half, c_pos_second_half = kahan_sum(to_add, pos_sum_second_half, c_pos_second_half)
        else:
            neg_sum_second_half, c_neg_second_half = kahan_sum(to_add, neg_sum_second_half, c_neg_second_half)
        one = _cast_int(1, dtype)
    return neg_sum_first_half + pos_sum_first_half + neg_sum_second_half + pos_sum_second_half

print(s_b(1000))
```

We will run a small routin to check the error between the two approaches.

```python
n = 200
partial_sums = np.zeros(n)
sums = np.zeros(n)

for i in range(1,n+1):
    partial_sums[i-1] = s_b(i)
    sums[i-1], _ = s(i)
    
```

```python
s(500)[0], s_b(500)
```

```python
errors = abs(partial_sums - sums) / abs(sums)
plt.plot(np.arange(1,n+1), errors)
plt.xlabel('n')
plt.ylabel('relative_error')
plt.title('Absolute Relative Error, Runinng Sums vs Same Sign Partial Sums')
plt.show()
```
